```python
# make the size of each marker equal to the data floating-point decimal values of the y-axis 
plt.scatter(x_axis, y_axis, s=y_axis)

# iterate through the y-axis data and multiply each data point by 3 
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

#the new y-axis list for the s parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# list comprehension technique to multiply each data point 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="blue", label='Boston', alpha=0.4, edgecolors="black", s = [i * 5 for i in y_axis])
# Create labels for the x and y axes.
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_xlim(4, 45)
# Create a title.
ax.set_title("PyBer Fare by month")

plt.gca().invert_yaxis()


# Add the legend.
ax.legend()
```
